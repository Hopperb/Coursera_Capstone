# Import Libs
Importing Libraries and reading the path to import to dataframe and began preprocessing of the df.


```python
#importing libraries
import pandas as pd
import numpy as np 
print("imported libraries")

#reading path to create dataframe
path = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
df = pd.read_html(path)
print(df)

#sepreating list of dfs and renaming columns in 1st df
#dropping "not assigned" from df
#dropping na from the second and third df
df1 = df[0]
df2 = df[1]
df3 = df[2]
print(df1, df2, df3)

df1 = df1.rename(index = str, columns={0:"Postcode", 1:"Borough", 2: "Neighbourhood" })
df1 = df1[df1.Borough != 'Not assigned']
df2 = df2.dropna()
df3 = df3.dropna()

#shaping the 3 dfs
print(df1.shape, df2.shape, df3.shape)
```

# New DF with Lat and Lon in col

```python
#name columns with dtypes
dataColumns = {"Postal Code":str, "Latitude":float, "Longitude":float, "Borough":str, "Neighbourhood":str}

#reading csv file into df
postal = pd.read_csv("https://cocl.us/Geospatial_data")

#renamed col for merge
postal = postal.rename(columns={"Postal Code": "Postcode"})

#merged dfs
postal = postal.merge(df1)
print(postal)
```
